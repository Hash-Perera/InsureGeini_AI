```python
from ultralytics import YOLO

#loading the model
model = YOLO('yolo11n-cls.pt')

#Data dir
DATA_DIR = '/content/dataset/data3a/'
#Train the model
result = model.train(
    data=DATA_DIR,
    epochs=100,
    imgsz=320,
    lr0=0.001,
    patience=20,
    weight_decay=0.0005,
    workers=8,
    batch=16
)
```

```python
#Inference
from ultralytics import YOLO
import os

# Loading model
model = YOLO('/content/drive/MyDrive/DamageClassification/train3/weights/best.pt')

# Function for inference
def classify_image(image_path):
    # Perform inference
    results = model(image_path)

    # Extract top prediction and confidence
    prediction = results[0].names
    confidences = results[0].probs

    # Get top-1 prediction using the 'top1' property of the Probs object
    top1_index = confidences.top1
    top1_class = prediction[top1_index]
    top1_confidence = confidences.top1conf

    print(f"Image: {image_path}")
    print(f"Predicted Class: {top1_class}, Confidence: {top1_confidence:.2f}")

    return top1_class, top1_confidence

# Input image
image_path = '/content/Test image 2.jpg'
classify_image(image_path)

```
